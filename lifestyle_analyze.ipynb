```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
dt_lifestyle = pd.read_csv('/content/Final_data.csv')

print(dt_lifestyle)
# print(dt_lifestyle['Session_Duration (hours)'])
```

```python
dt_lifestyle.info()
```

```python
dt_lifestyle.describe()
```

#Univariate

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Gender", data=dt_lifestyle, hue='Gender')
plt.title("Count of Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(dt_lifestyle["Age"], bins=30, kde=True, color="C0")
plt.axvline(dt_lifestyle["Age"].mean(), color="red", linestyle="--", label=f"mean={dt_lifestyle['Age'].mean():.1f}")
plt.axvline(dt_lifestyle["Age"].median(), color="green", linestyle="-.", label=f"median={dt_lifestyle['Age'].median():.1f}")
plt.title("Age distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(dt_lifestyle["BMI"], bins=30, kde=True, color="C0")
plt.axvline(dt_lifestyle["BMI"].mean(), color="red", linestyle="--", label=f"mean={dt_lifestyle['BMI'].mean():.1f}")
plt.axvline(dt_lifestyle["BMI"].median(), color="green", linestyle="-.", label=f"median={dt_lifestyle['BMI'].median():.1f}")
plt.title("BMI distribution")
plt.xlabel("BMI")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
def bmi_cat(bmi):
    if bmi < 18.5: return "underweight"
    if bmi < 25: return "normal"
    if bmi < 30: return "overweight"
    return "obese"
dt_lifestyle["BMI_cat"] = dt_lifestyle["BMI"].apply(bmi_cat).astype("category")
# dt_lifestyle["BMI_cat"]

plt.figure(figsize=(8,5))
sns.countplot(x="BMI_cat", data=dt_lifestyle, hue='BMI_cat')
plt.title("Category of BMI")
plt.xlabel("BMI")
plt.ylabel("Count")
plt.show()
```

#Bivariate

```python
# dt_lifestyle['Workout_Frequency (days/week)'] = dt_lifestyle['Workout_Frequency (days/week)'].astype(int)
# dt_lifestyle['Calories_Burned'] = dt_lifestyle['Calories_Burned'].astype(int)

sns.regplot(
    data=dt_lifestyle,
    x='Workout_Frequency (days/week)',
    y='Calories_Burned',
    scatter_kws={'color':'skyblue'},
    line_kws={'color':'red'}
)
plt.title('Workout Frequency and Calories Burned Relation')
plt.xlabel('Workout Frequency (days/week)')
plt.ylabel('Calories Burned')
plt.show()
```

```python
# dt_lifestyle['Session_Duration (hours)'] = dt_lifestyle['Session_Duration (hours)'].astype(int)
sns.regplot(
    data=dt_lifestyle,
    x='Session_Duration (hours)',
    y='Calories_Burned',
    scatter_kws={'color':'yellow'},
    line_kws={'color':'red'}
)
plt.title('Session_Duration and Calories Burned Relation')
plt.xlabel('Session_Duration (hours)')
plt.ylabel('Calories Burned')
plt.show()
```

```python
# workout_countbygender = dt_lifestyle.groupby(['Gender','Workout_Type']).size().reset_index(name='workout_bygender')

chart = sns.barplot(x='Workout_Type', y='Calories_Burned', data=dt_lifestyle)
chart.set_title('Workout Type and Calories Burned Relation')
plt.xlabel('Workout Type')
plt.ylabel('Calories_Burned')
plt.tight_layout()
plt.show()
```

```python
chart = sns.scatterplot(x='Age', y='Avg_BPM', data=dt_lifestyle, hue='Workout_Type')
chart.set_title('Avarage BPM by Age')
plt.xlabel('Age')
plt.ylabel('Average BPM')
plt.tight_layout()
plt.show()
```

```python
chart = sns.scatterplot(x='protein_per_kg', y='lean_mass_kg', data=dt_lifestyle, hue='BMI_cat')
chart.set_title('Avarage BPM by Age')
plt.xlabel('protein')
plt.ylabel('lean mass')
plt.tight_layout()
plt.show()
```

#Multivariate

```python
dt_encoded = dt_lifestyle[['Calories_Burned', 'Workout_Type', 'Session_Duration (hours)', 'Workout_Frequency (days/week)','Fat_Percentage', 'BMI', 'lean_mass_kg', 'Resting_BPM' ,'Max_BPM', 'Age','Experience_Level']].copy()
dt_encoded = pd.get_dummies(dt_encoded, drop_first=True)

sns.heatmap(dt_encoded.corr(), annot=False, cmap='coolwarm')
plt.title('Heatmap Korelasi (termasuk kolom kategori)')
plt.show()
```
